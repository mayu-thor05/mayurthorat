## Functions and Methods

```python
# Functions are the globally accessible block of reusable code. We can deifne them outside a class using 'def' keyword 

def calculate_rec_area(width, height):
    return width * height

#calls independently by name
rectangle_area = calculate_rec_area(5,10)
print(rectangle_area)
```

```python
# Python Methods are define inside the class block. It defines actions and behaviour an object can perform

class Rectangle:
    def __init__(self,width,height):
        self.height = height
        self.width = width
    #This is a method as it is inside the class and takes self object
    def calculate_area(self):
        return self.height * self.width

# Must intantiate the object first
my_rect = Rectangle (5,10)

#Called object using dot notation

rectangle_area = my_rect.calculate_area()
print(rectangle_area)

    
```

### OOPs organizes code around objects that bundle data and behviour together while the Functional Programming (FP) organizes code around pure, independent mathematical functions. OOPs focuses on managing mutable states via inheritance and encapsulation, while the FP relies on data immutablity and mathematical transformation pipelines

### Array Practice Problems

```python
# Largest No from the Array

n = int(input())

arr = list(map(int, input().split()))

print(arr)

largest = arr[0]

for i in range(1,n):
    if arr[i] > largest:
        largest = arr[i]
print(largest)        
```

```python
# Smallest No from the Array
n=int(input())

arr=list(map(int,input().split()))

print(arr)

smallest = arr[0]

for i in range(1,n):
    if arr[i] < smallest:
        smallest = arr[i]
print(smallest)        
```

```python
# Sum of All elements of an Array
n = int(input())

arr = list(map(int, input().split()))

print(arr)
addition = 0
for num in arr:
    addition +=num
print(addition)
```

```python
# Average of the Array

n = int(input())

arr = list(map(int,input().split()))

total = 0
for num in arr:
    total +=num
average = total/n    
print(average)    
```

```python
# Count Odd and Even Numbers
n = int(input())
arr = list(map(int,input().split()))
print(arr)
counteven=0
countodd=0
for i in range(0,n):
    if arr[i]%2 == 0:
        counteven += 1
    else:
        countodd += 1
print("Even No:",counteven)
print("Odd No:",countodd)
        
```

```python
#Reverse an Array

n = int(input())
arr = list(map(int,input().split()))

for i in range(n-1, -1, -1):
    print(arr[i], end="")
```
